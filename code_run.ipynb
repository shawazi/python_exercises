```python
def string_counter(string):
    vowels = 0
    consonants = 0
    cons_list = [ "b", "c", "d", "f", "g", "h", "j", "k", "l", "m", "n", "p", "q", "r", "s", "t", "v", "w", "x", "z" ]
    
    for letter in string.lower():
        match letter:
            case "a":
                vowels += 1
            case "e":
                vowels += 1
            case "i":
                vowels += 1
            case "o":
                vowels += 1
            case "u":
                vowels += 1
            case "y":
                vowels += 1
            case cons if cons in cons_list:
                consonants += 1
    print("Vowel count: ", vowels)
    print("Consonant count: ", consonants)
    
string_counter("abcdefghijklmnopqrstuvwxyz __389(*$&#@(@)&$#*&!!@#@__  ABCDEFGHIJKLMNOPQRSTUVWXYZ")
            
    
```

```python
dict1 = {'Apple': 15, 'Orange': 35, 'Melon': 20, 'Banana': 50, 'Pear': 40}  

def dict_doubler(dict):
    dict2 = {}
    for key in dict:
        # print(key, value)
        dict2[key] = dict1[key] * 2
    print(dict2)
    
dict_doubler(dict1)
```

```python
def repeater():
    word = input("Enter a word: ")
    sep = input("Enter a separator: ")
    num_rep = int(input("Enter the number of repititions: "))
    
    for i in range(num_rep):
        if i == num_rep -1:
            print(word, end="")
        else:
            print(word, end=sep)
         
repeater()
```
